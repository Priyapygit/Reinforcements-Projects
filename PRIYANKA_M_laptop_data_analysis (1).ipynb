```python
#Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
laptop_clean = laptop[['Ram','Price']].dropna()

X = laptop_clean[['Ram']]
y = laptop_clean['Price']

model = LinearRegression()
model.fit(X, y)

slope = model.coef_[0]
intercept = model.intercept_

print(f"Regression Equation: Price = {intercept:.2f} + {slope:.2f} * RAM")
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Ram', y='Price', data=laptop_clean, label='Data')
x_vals = np.linspace(laptop_clean['Ram'].min(), laptop_clean['Ram'].max(), 100)
y_vals = intercept + slope * x_vals
plt.plot(x_vals, y_vals, color='red', label='Regression Line')
plt.title('RAM vs Price with Regression Line')
plt.xlabel('RAM (GB)')
plt.ylabel('Price')
plt.legend()
plt.show()

print(f"Slope: {slope:.2f}")
```

```python
#Interpretation: The slope is positive, adding more RAM significantly contributes to higher prices on average
```

Lightweight premium laptops command higher prices,
